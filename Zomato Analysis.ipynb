```python
import pandas as pd
import numpy as np

class RestaurantAnalyzer:
   
    def __init__(self, filepath):
        
        self.filepath = filepath
        self.df = None  # We initialize the dataframe as None. It will be loaded by our method.
        print("RestaurantAnalyzer object created.")

    def load_data(self):
        
        try:
            self.df = pd.read_csv(self.filepath)
            print(f"Data loaded successfully. Shape: {self.df.shape}")
        except FileNotFoundError:
            print(f"Error: The file at {self.filepath} was not found.")
            print("Please check the file path.")
            self.df = None 
        
        return self.df
```

```python
DATA_PATH = r"C:\python full stack\Likith\Zomato Analysis Project\zomato.csv"

analyzer = RestaurantAnalyzer(filepath=DATA_PATH)

df = analyzer.load_data()
```

```python
# Create a copy to keep the original data safe
df_cleaned = df.copy()
```

```python
# Drop the columns that are not required for our analysis
df_cleaned = df_cleaned.drop(columns=['url', 'dish_liked', 'phone'])

# Let's check the result
print("Columns after dropping:")
print(df_cleaned.columns)
print(f"\nNew shape of the data: {df_cleaned.shape}")
```

```python

# Rename columns for easier access
df_cleaned.rename(columns={
    'approx_cost(for two people)': 'approx_cost',
    'listed_in(type)': 'meal_type',
    'listed_in(city)': 'area'
}, inplace=True)

# Check the new column names
print("Renamed columns:")
print(df_cleaned.columns)
```

```python
# First, let's see all the unique values to understand the problem
print("Unique values in 'rate' column before cleaning:")
print(df_cleaned['rate'].unique())
```

```python
# Let's define a rule (as a small, one-line function) to clean these values
# We use a 'lambda' function here, which is just a way to write a quick, anonymous function
def clean_rate(rate_str):
    if isinstance(rate_str, str):
        if "NEW" in rate_str or "-" in rate_str:
            return np.nan  # Return a missing value for 'NEW' and '-'
        else:
            return float(rate_str.split('/')[0]) # Split "4.1/5" and take the "4.1" part
    return rate_str # Keep it as is if it's already a number or NaN
```

```python
# Apply this rule to the entire 'rate' column
df_cleaned['rate'] = df_cleaned['rate'].apply(clean_rate)
```

```python
print("\nUnique values in 'rate' column after cleaning:")
print(df_cleaned['rate'].unique())
print(f"\nData type of 'rate' column now: {df_cleaned['rate'].dtype}")
```

```python
# Check the unique values first
print("Unique values in 'approx_cost' column before cleaning:")
print(df_cleaned['approx_cost'].unique()[:20]) # Show first 20 unique values
```

```python
# Remove commas and convert to numeric
df_cleaned['approx_cost'] = df_cleaned['approx_cost'].str.replace(',', '', regex=False)
df_cleaned['approx_cost'] = pd.to_numeric(df_cleaned['approx_cost'])
```

```python
print("\nUnique values in 'approx_cost' column after cleaning:")
print(df_cleaned['approx_cost'].unique()[:20])
print(f"\nData type of 'approx_cost' column now: {df_cleaned['approx_cost'].dtype}")
```

```python
df_cleaned.info()
```

```python
# For numerical columns like 'rate' and 'approx_cost', filling with the median is a good strategy
# because it is not affected by extreme outliers.
rate_median = df_cleaned['rate'].median()
df_cleaned['rate'] = df_cleaned['rate'].fillna(rate_median)

cost_median = df_cleaned['approx_cost'].median()
df_cleaned['approx_cost'] = df_cleaned['approx_cost'].fillna(cost_median)
```

```python
# For categorical columns, filling with the mode (most frequent value) is the best approach.
location_mode = df_cleaned['location'].mode()[0]
df_cleaned['location'] = df_cleaned['location'].fillna(location_mode)

rest_type_mode = df_cleaned['rest_type'].mode()[0]
df_cleaned['rest_type'] = df_cleaned['rest_type'].fillna(rest_type_mode)

cuisines_mode = df_cleaned['cuisines'].mode()[0]
df_cleaned['cuisines'] = df_cleaned['cuisines'].fillna(cuisines_mode)

print("Missing values have been handled.")

```

```python
print("Final Data Information:")
df_cleaned.info()
```

```python
print("\n\nFinal Count of Missing Values:")
print(df_cleaned.isnull().sum())
```

```python
#Exploratory Data Analysis (EDA).
import matplotlib.pyplot as plt
import seaborn as sns

# Set a style for all our plots
sns.set_style('darkgrid')
# Set the default figure size
plt.rcParams['figure.figsize'] = (10, 7)

print("Plotting libraries imported and styles set.")
```

```python
# Create a histogram of the 'rate' column
sns.histplot(df_cleaned['rate'], bins=20, kde=True)
plt.title('Distribution of Restaurant Ratings', fontsize=14)
plt.xlabel('Rating (out of 5)', fontsize=14)
plt.ylabel('Number of Restaurants', fontsize=14)
plt.show()
```

```python
# Create a count plot for the 'online_order' column
sns.countplot(x='online_order', data=df_cleaned)
plt.title('Restaurants Offering Online Order Facility', fontsize=14)
plt.xlabel('Offers Online Order?', fontsize=14)
plt.ylabel('Number of Restaurants', fontsize=14)
plt.show()
```

```python
# Let's compare the ratings of restaurants with and without online ordering
sns.boxplot(x='online_order', y='rate', data=df_cleaned)
plt.title('Rating vs. Online Order Availability', fontsize=14)
plt.xlabel('Offers Online Order?', fontsize=14)
plt.ylabel('Rating (out of 5)', fontsize=14)
plt.show()
```

```python
# Analyze the relationship between approximate cost and rating
# A jointplot with hexbin is great for visualizing dense data
sns.jointplot(x='approx_cost', y='rate', data=df_cleaned, kind='hex', cmap='viridis_r')
plt.suptitle('Density of Cost vs. Rating', y=1.02, fontsize=14 ) # Add a main title
plt.show()
```

```python
# Analyze the impact of table booking on ratings
sns.boxplot(x='book_table', y='rate', data=df_cleaned)
plt.title('Rating vs. Table Booking Availability', fontsize=14)
plt.xlabel('Offers Table Booking?', fontsize=14)
plt.ylabel('Rating (out of 5)', fontsize=14)
plt.show()
```

```python
# To make the plot readable, let's increase the figure size
plt.figure(figsize=(12, 6))
# Analyze ratings across different restaurant types
sns.boxplot(x='rest_type', y='rate', data=df_cleaned)
plt.title('Rating vs. Restaurant Type', fontsize=14)
plt.xlabel('Restaurant Type', fontsize=14)
plt.ylabel('Rating (out of 5)', fontsize=14)
plt.xticks(rotation=90) # Rotate the x-axis labels to prevent overlap
plt.show()
```

```python
# Step 1: Find the top 15 areas with the most restaurants
top_areas = df_cleaned['area'].value_counts().nlargest(15).index
```

```python
# Step 2: Filter the DataFrame to include only these top areas
df_top_areas = df_cleaned[df_cleaned['area'].isin(top_areas)]
```

```python

# Step 3: Create the plot
plt.figure(figsize=(15, 6))
sns.boxplot(x='area', y='rate', data=df_top_areas, palette='viridis')
plt.title('Ratings in Top 15 Areas', fontsize=14)
plt.xlabel('Area', fontsize=14)
plt.ylabel('Rating (out of 5)', fontsize=14)
plt.xticks(rotation=45)
plt.show()
```

```python
# Find the top 10 most common cuisines
top_cuisines = df_cleaned['cuisines'].value_counts().nlargest(10).index

# Filter the DataFrame to include only these top cuisines
df_top_cuisines = df_cleaned[df_cleaned['cuisines'].isin(top_cuisines)]

# Let's plot their ratings
plt.figure(figsize=(15, 6))
sns.boxplot(x='cuisines', y='rate', data=df_top_cuisines, palette='inferno')
plt.title('Ratings of Top 10 Most Popular Cuisines', fontsize=14)
plt.xlabel('Cuisine', fontsize=14)
plt.ylabel('Rating (out of 5)', fontsize=14)
plt.xticks(rotation=45)
plt.show()
```
